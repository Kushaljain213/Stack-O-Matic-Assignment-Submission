```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Loading Dataset
df = pd.read_csv('extended_salary_data.csv')
print(df)
```

```python
#Scatter Plot
plt.scatter(df['YearsExperience'],df['Salary'])
plt.title("Scatter Plot")
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Mean , Median and Variance
print('Mean of Years Experienced is :',df['YearsExperience'].mean())
print('Median of Years Experienced is :',df['YearsExperience'].median())
print('Variance of Years Experienced is :',df['YearsExperience'].var())



print('Mean of Salary is :',df['Salary'].mean())
print('Median of Salary is :',df['Salary'].median())
print('Variance of Salary is :',df['Salary'].var())



```

```python
#train_test_spilit
from sklearn.model_selection import train_test_split
x = df[['YearsExperience']]
y = df['Salary']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 42)

```

```python
#Linear Regression on Training set
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)

slope = model.coef_[0]
intercept = model.intercept_
```

```python
#Coefficients of the linear model
print(slope)
print(intercept)
```

```python
#Making Predictions of testing set
y_predict = model.predict(x_test)
print(y_predict)
```

```python
#MAE and MAS
from sklearn.metrics import mean_absolute_error, mean_squared_error

print("Mean Absolute Error (MAE) is :",mean_absolute_error(y_test,y_predict))
print("Mean Squared Error (MSE) is : ",mean_squared_error(y_test,y_predict))
```

```python
years_range = np.linspace(x.min(), x.max(),100).reshape(-1,1)
salary_predictions = model.predict(years_range)
plt.scatter(x,y, color= 'blue',label = 'Data Points')
plt.plot(years_range , salary_predictions , color = 'red', label = 'Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Regression line over scatter plot')
plt.legend()
```
